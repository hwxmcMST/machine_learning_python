```python
# Define a function to determine the borrow time
def calcu_loan():
    # Capture user input of borrower type
    borrower = input("Enter the type of borrower (Student/Faculty/Other):").upper()
    # Validate borrower type
    while borrower != "STUDENT" and borrower != "FACULTY" and borrower != "OTHER":
        borrower = input("Enter the type of borrower (Student/Faculty/Other):").upper()
        
    # Capture material type
    material = input("Enter the type of materials to borrow (Book/CD):").upper()
    # Validate material type
    while material != "BOOK" and material != "CD":
        material = input("Enter the type of materials to borrow (Book/CD):").upper()
        
    # Print out load period for each combination of borrower type and material type
    if(borrower == "STUDENT" and material == "BOOK"): 
        print("You can borrow books for 4 weeks.")
        
    if(borrower == "STUDENT" and material == "CD"):
        print("You can borrow CDs for 1 week.")
    
    if(borrower == "FACULTY" and material == "BOOK"): 
        print("You can borrow books for 6 months.")
    
    if(borrower == "FACULTY" and material == "CD"): 
        print("You can borrow CDs for 2 weeks.")
    
    if(borrower == "OTHER" and material == "BOOK"):
        print("You can borrow books for 2 weeks.")
    
    if(borrower == "OTHER" and material == "CD"):
        print("You can borrow CDs for 1 week.")
```

```python
# Run the function
calcu_loan()
```
