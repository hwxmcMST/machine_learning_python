# Task: String and Dataset Manipulation (10 points)

(1) Create a Pandas Series named contact that contains the following string elements:

- '101 Fulton Hall, 301 W. 14th St., Rolla, MO, 65409, (573) 341-7216, bit@mst.edu'
- '107 IDE Building, 1215 N. Pine St., Rolla, MO 65409, (573) 341-7525, ithelp@mst.edu'
- '400 W. 14th Street, Rolla, MO 65409, (573) 341-4008, library@mst.edu'

```python
import pandas as pd
```

```python
contact = pd.Series(['101 Fulton Hall, 301 W. 14th St., Rolla, MO, 65409, (573) 341-7216, bit@mst.edu',
                    '107 IDE Building, 1215 N. Pine St., Rolla, MO 65409, (573) 341-7525, ithelp@mst.edu',
                    '400 W. 14th Street, Rolla, MO 65409, (573) 341-4008, library@mst.edu'])
```

```python
contact
```

(2) Extract zip code from the contact object.

```python
# Zip code is a substring that simply contains 5 decimal digits.
zip = contact.str.extract('(\d{5})', expand = True)
zip
```

(3) Extract MST email address from the contact object.

```python
email = contact.str.extract('(\w+@mst.edu)', expand = True)
email
```

(4) Create a newcontact object which contains the original contact object and the zip code and email addresses extracted in (2) and (3).

Rename the columns of this newcontact object as 'Contact', 'Zip Code', and 'Email'.

Print the content of the newcontact object.

```python
newcontact = pd.concat([contact,zip,email], axis = 1)
newcontact.columns = ['Concact', 'Zip Code', 'Email']
newcontact
```

(5) Select observations in the newcontact object that contains the substring '14th' in the Contact column.

```python
newcontact[newcontact.Concact.str.contains('14th')]
```
