# Task 2: Loop (4 points)

2.1 Define a function to convert a score value to a letter grade. 

```python
# Define the function 
def score_2_grade(score):
    if score >= 90:
        grade = 'A'
    elif score >= 80:
        grade = 'B'
    elif score >= 70:
        grade = 'C'
    elif score >= 60:
        grade = 'D'
    else:
        grade = 'F'
    return grade
```

2.2 Use a for loop to calculate letter grade for each value in the score array. Store the letter grade in an array object called "grade".

```python
# Create a grade array with no element
grade = np.array([])    
```

```python
for i in score:
    grade = np.append(grade,score_2_grade(i))  
```

2.3 Show the content of the grade array

```python
grade
```

2.4. Combine the score and grade arrays into a 7 by 2 matrix. The 1st column shows the score and the 2nd column shows the letter grade. Show the content of this matrix.

```python
np.concatenate((score.reshape(7,1), grade.reshape(7,1)), axis = 1)
```
